#### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

#### Importing the dataset

```python
## Load the dataset
df = pd.read_csv('Churn_modelling.csv')
df.head()
```

#### Data Preprocessing

```python
## Data Preprocessing
## Drop Irrelevant Features
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
df
```

```python
## Encoding Categorical Variables
## Label Encoding 'Gender' Column
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
## Encoding Categorical Variables
## One Hot Encoding 'Geography' Column
from sklearn.preprocessing import OneHotEncoder
ohe_geo = OneHotEncoder()
geo_encoder = ohe_geo.fit_transform(df[['Geography']])
geo_encoder
```

```python
ohe_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoder.toarray()
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(), columns = ohe_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
df.head()
```

```python
## Combine One Hot Encoded Data with the original data
df = pd.concat([df.drop('Geography', axis = 1), geo_encoded_df], axis = 1)
df.head()
```

```python
## Save the Encoders
with open('label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('ohe_geo.pkl', 'wb') as file:
    pickle.dump(ohe_geo,file)
```

```python
## Divide the Dataset into Dependent and Independent Features
X = df.drop('Exited', axis = 1)
y = df['Exited']
```

#### Train-Test-Split

```python
## Splitting the the Data in Training and Test Set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

#### Scaling the Data

```python
## Scaling the Data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Saving the Scaler
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler,file)
```

### ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
## Building our Model
model = Sequential([
    Dense(units = 64, activation = 'relu', input_shape = (X_train.shape[1],)),  ## Hidden Layer 1
    Dense(units = 32, activation = 'relu'),  ## Hidden Layer 2
    Dense(units = 1, activation = 'sigmoid')  ## Output Layer
])
```

```python
model.summary()
```

```python
## Compiling the Model
model.compile(optimizer = 'adam', metrics = ['accuracy'], loss = 'binary_crossentropy')

'''
opt = tf.keras.optimizers.Adam(learning_rate = 0.01)  ## Adam Optimizer with custom Learning Rate
loss = tf.keras.losses.BinaryCrossentropy()
model.compile(optimizer = opt, metrics = ['accuracy'], loss = loss)
'''
```

```python
## Setup the Tensorboard
log_dir = 'logs/fit'
tensorflow_callbacks = TensorBoard(log_dir = log_dir, histogram_freq = 1)
```

```python
## Setup Early Stopping
early_stopping_callback = EarlyStopping(monitor = 'val_loss', patience = 10, restore_best_weights = True)
```

```python
## Training the Model
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 10, callbacks = [tensorflow_callbacks, early_stopping_callback])
```

```python
model.save('model.h5')
```
